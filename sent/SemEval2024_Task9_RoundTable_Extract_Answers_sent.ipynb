```python
import pandas as pd
import re

# Assuming your DataFrame is named 'df'
# If it's not loaded, you can load it using pd.read_csv() or another appropriate method

# Define a regular expression pattern to match the desired format
pattern = r"Answer: (\S+)\nConfidence: (\d+)%"
# pattern = r"Answer: (\S+) (.*)?\nConfidence: (\d+)%"

def extract_answer(val):
    # Create empty lists to store extracted answers and confidences
    answers = []
    confidences = []

    # Iterate through the val column
    for value in df[val]:
        match = re.search(pattern, value)

        if match:
            # Extract answer and confidence from the matched groups
            answer = match.group(1)
            confidence = int(match.group(2))

            # Append to the lists
            answers.append(answer)
            confidences.append(confidence)
        else:
            # Print rows that do not follow the specified format
            print("Invalid format:", value)

    # Create new columns in the DataFrame for extracted answers and confidences
    # df['Extracted_Answer'] = answers
    # df['Extracted_Confidence'] = confidences

    # # Print the DataFrame with extracted information
    # print(df)
    return answers, confidences
```

```python
answers, confidences = extract_answer('Round_1_Model_ChatGPT')
```
